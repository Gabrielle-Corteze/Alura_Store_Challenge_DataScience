# 4. Produtos Mais e Menos Vendidos

Loja 1


```python
Produtos = loja['Produto'].value_counts(ascending = False)
Produtos_mais_vendidos = Produtos.head(3)
Produtos_mais_vendidos.plot(kind='bar', figsize = (8, 4), color='green')
```

```python
Produtos_menos_vendidos = Produtos.tail(3)
Produtos_menos_vendidos.plot(kind='bar', figsize = (8, 4), color='red')
```

```python
def comparar_vendas(Produtos_mais_vendidos, Produtos_menos_vendidos):
  comparacao_vendas= pd.concat([Produtos_mais_vendidos, Produtos_menos_vendidos]).sort_values()

  plt.figure(figsize=(8, 4))
  cores = ['red'] * 3 + ['green'] * 3
  comparacao_vendas.plot(kind='barh', color=cores)

  plt.title('Mais Vendidos X Menos Vendidos')
  plt.xlabel('Quantidade de Vendas')
  plt.ylabel('Produtos')
  plt.grid(axis='x', linestyle='--', alpha=0.6)

  plt.tight_layout()
  plt.show()

comparar_vendas(Produtos_mais_vendidos, Produtos_menos_vendidos)
```

# 📑 Relatório Estratégico de Performance Comercial

## 1. Introdução
Este documento apresenta a síntese dos indicadores de desempenho das quatro unidades comerciais sob análise. O objetivo é fundamentar a recomendação de venda de uma das unidades, visando a otimização de recursos e o foco em operações de maior rentabilidade.

---

## 2. Análise dos Indicadores de Desempenho

Abaixo, detalhamos os fatores críticos que basearam esta decisão:

* **Faturamento Total:** A unidade **Loja 4** registrou o menor volume financeiro do grupo, apresentando uma defasagem considerável em relação à Loja 1, que lidera o ranking.
* **Eficiência por Categoria:** Foi identificado que a **Loja 4** é a única operação que se encontra abaixo da média geral de desempenho por categoria de produtos.
* **Satisfação do Cliente:** As médias de avaliações apontam que as **Lojas 2 e 3** possuem maior fidelidade e satisfação do público, enquanto a **Loja 4** apresenta notas inferiores (média próxima a 4.0), demandando maior investimento em CRM.
* **Logística e Fretes:** Apesar de a **Loja 4** possuir o menor custo médio de frete, os dados comprovam que este benefício não se traduziu em aumento de vendas ou competitividade.
* **Volume de Vendas:** No ranking de produtos mais vendidos, a **Loja 4** ocupa posições secundárias, demonstrando baixa tração comercial comparada às unidades de topo.

---

## 3. Justificativa da Recomendação

Após o cruzamento de dados, a **Loja 4** é identificada como a melhor candidata para desinvestimento. A justificativa baseia-se nos seguintes pontos:

1.  **Baixo ROI (Retorno sobre Investimento):** É a loja com menor contribuição para o lucro líquido total.
2.  **Inconsistência de Desempenho:** Apresenta variabilidade nos indicadores sem demonstrar tendência de crescimento sustentável.
3.  **Vantagem Competitiva Inexplorada:** A unidade falhou em aproveitar sua estrutura de baixo custo de frete para liderar em volume de vendas.

---

## 4. Conclusão Final

**Recomendação:** Venda da **Loja 4**.

A análise evidencia que a continuidade da operação da Loja 4 gera um custo de oportunidade elevado. Recomenda-se a alienação desta unidade para que o capital e o esforço de gestão sejam redirecionados às Lojas 1, 2 e 3, que demonstram indicadores de saúde financeira e aceitação de mercado superiores.
